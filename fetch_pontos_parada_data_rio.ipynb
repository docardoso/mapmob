```python
import urllib.request
import os.path
import json
import pandas as pd
import time
import datetime
```

```python
def fetch_data_page(url):
    resource = urllib.request.urlopen(url)
    content =  resource.read().decode(resource.headers.get_content_charset())
    return json.loads(content)
```

```python
# https://www.data.rio/documents/PCRJ::pontos-de-parada-de-%C3%B4nibus-da-cidade/about
# https://jeap.rio.rj.gov.br/MOB/get_stops_details.rule?sys=MOB&INDICE=0

data_page = fetch_data_page('https://jeap.rio.rj.gov.br/MOB/get_stops_details.rule?sys=MOB&INDICE=0')
```

```python
data_page.keys()
```

```python
all_data = []
```

```python
while True:
    all_data.extend(data_page['data'])
    print(data_page['next'])
    
    try:
        data_page = fetch_data_page(data_page['next'])
    except ValueError:
        break
```

```python
all_data = pd.DataFrame(all_data)
all_data
```

```python
all_data.to_csv('pontos_parada_data_rio_' + time.strftime('%Y-%m-%d-%H-%M-%S') +'.csv')
```
